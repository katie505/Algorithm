```python
n = 15
```

```python
# My answer
from typing import List
class Solution1:
    def fizzBuzz(self, n: int) -> List[str]:
        answer = []
        for i in range(1, n+1):
            if i % 3 == 0 & i % 5 == 0:
                answer.append('FizzBuzz')
            elif i % 3 == 0:
                answer.append('Fizz')
            elif i % 5 == 0:
                answer.append('Buzz')
            else:
                answer.append(str(i))
        
        return answer
```

```python
answer = Solution1()
answer.fizzBuzz(n)
```
