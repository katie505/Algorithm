```python
nums = [1,2,3,4]
```

```python
# My answer
from typing import List
class Solution1:
    def runningSum(self, nums: List[int]) -> List[int]:
        runningSum = []
        sum = 0
        
        for i in range(len(nums)):
            sum += nums[i]
            runningSum.append(sum)
            
        return runningSum
```

```python
answer = Solution1()
answer.runningSum(nums)
```
