```python
accounts = [[1,2,3],[3,2,1]]
```

```python
# My answer
class Solution1:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        max_sum = []
        for i in range(len(accounts)):
            sum = 0
            for j in range(len(accounts[i])):
                sum += accounts[i][j]
            max_sum.append(sum)
        return max(max_sum)
```

```python
answer = Solution1()
answer.maximumWealth(accounts)
```

```python
# 2중 for loop를 활용한 정답 (공부를 위한 코드)
class Solution2:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        Max = -1
        for i in range( len( accounts) ):
            currentSum = 0
            for j in range(len(accounts[i])):
                currentSum += accounts[i][j]
                if Max < currentSum:
                    Max = currentSum
        return Max
```

```python
answer = Solution2()
answer.maximumWealth(accounts)
```

```python
# sum 함수를 이용한 정답
class Solution3:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        Max = -1
        for i in range( len( accounts) ):
            if Max < sum(accounts[i]):
                Max = sum(accounts[i])
        return Max
```

```python
answer = Solution3()
answer.maximumWealth(accounts)
```
