```python
ransomNote1 = "aa"
magazine1 = "aab"
ransomNote2 = "aa"
magazine2 = "ab"
```

```python
# My answer
from collections import Counter
class Solution1:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        dict_ran = Counter(ransomNote)
        dict_maz = Counter(magazine)

        for key, value in dict_ran.items():
            if key not in dict_maz or value > dict_maz[key]:
                return False
            
        return True
```

```python
answer = Solution1()
answer.canConstruct(ransomNote1, magazine1)
```

```python
answer.canConstruct(ransomNote2, magazine2)
```

```python
from collections import Counter
class Solution2:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
      ransomNoteList = list(ransomNote)
      magazineList = list(magazine)

      Answer = True
      for char in ransomNoteList:
        if char not in magazineList:
          Answer = False
          break
        else:
          magazineList.pop(magazineList.index(char))
      return Answer
```

```python
answer = Solution2()
answer.canConstruct(ransomNote1, magazine1)
```

```python
answer.canConstruct(ransomNote2, magazine2)
```
