```python
num = 14
```

```python
# My answer
from typing import List
class Solution1:
    def numberOfSteps(self, num: int) -> int:
        count = 0
        while num != 0:
            if num % 2 == 0:
                num /= 2
            else:
                num -= 1
            count += 1
        return count
```

```python
answer = Solution1()
answer.numberOfSteps(num)
```
